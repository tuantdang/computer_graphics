### **Catersian Coordinate System**
- Represented by two perpendicular axes $\vec{Ox}, \vec{Oy}$
- Where $O$ is the origin

### **Points**
- Point is the simplest concept of geometry
- A point is represented in 2D Cartisian Coodinates as $P=[x_p,y_p]^T$

### **Vectors**
- A vector $\vec{v}$ can be obtained from two points $P$ and $Q$ to form vector $\vec{v} = \vec{PQ}$ which includes **two important features**:
  - Direction: from $P$ to $Q$
  - Length (magnitude): $||\vec{v}||$
- A vector $\vec{PQ}$ is calcuated by subjected two points: $\vec{v} = \vec{PQ} = Q - P = [x_q-x_p, y_q-y_p]^T = [x_v, x_y]$ where $x_v=x_q-x_p, y_v=y_q-y_p$
- Note that:
  - Vector $\vec{QP}=P-Q = - (P-Q) = - \vec{PQ}$: $\vec{QP}$ has inverse direction to $\vec{PQ}$, but same length.
  - A point can reprent as an vector in Cartersian Coodinate:
    - $P=\vec{OP}$, $Q=\vec{OQ}$
    - $\vec{PQ} = Q - P = \vec{OQ} - \vec{OP} =  \vec{PO} + \vec{OQ}$

#### **Third point rule**
- $âˆ€A \in $ Cartersian Coodinates: $\vec{PQ} = \vec{PA} + \vec{AQ}$

#### **Dot Product**
-  Dot product bewtween two vectors are defined: $\vec{v} \cdot \vec{w} = ||\vec{v}||||\vec{w}||cos\alpha $. The result is a number or **scalar**.
- In 2D Cartersian: $\vec{v} \cdot \vec{w} = v_x*w_x +  v_y*w_y$
- In n-dimensions: $\vec{v} \cdot \vec{w} = \underset{i=1}{\overset{n}{\sum}}v_i*w_i $
- **Magnitude** of projection $\vec{v}$ on $\vec{w}$ is calculated: $||\vec{v}||cos\alpha = \frac{\vec{v} \cdot \vec{w}}{||\vec{w}||}$
- **Magnitude** of projection $\vec{w}$ on $\vec{v}$ is calculated: $||\vec{w}||cos\alpha = \frac{\vec{v} \cdot \vec{w}}{||\vec{v}||}$
- **Image** of projection $\vec{v}$ on $\vec{w}$ is calculated: $\vec{v}_{proj} = \frac{\vec{v} \cdot \vec{w}}{||\vec{w}||} \frac{\vec{w}}{||\vec{w}||}$
- **Image** of projection $\vec{w}$ on $\vec{v}$ is calculated: $\vec{w}_{proj} = \frac{\vec{v} \cdot \vec{w}}{||\vec{v}||} \frac{\vec{v}}{||\vec{v}||}$




```python
import numpy as np
from numpy import array
import matplotlib.pyplot as plt
nrow, ncol = 1, 1
plt.rcParams["figure.figsize"] = (4*ncol, 4*nrow)
fig, ax = plt.subplots(nrow, ncol)
ax.set_xlim([0, 10]); ax.set_ylim([0, 10])

P = array([2, 4])
Q = array([6, 6])
K = array([8, 4])
v = Q-P
w = K-P

len_v = np.sqrt(v[0]**2 + v[1]**2)
len_w = np.sqrt(w[0]**2 + w[1]**2)
dot_vw = v[0]*w[0] + v[1]*w[1] # or use numpy as np.dot(v, w)
unit_v = (v/len_v)
unit_w = (w/len_w)

v_proj = (dot_vw/len_w)*unit_w
w_proj = (dot_vw/len_v)*unit_v

# Visualize
head_size = 0.2
ax.text(P[0], P[1]+0.5, 'P', c='blue')
ax.text(Q[0], Q[1]+0.5, 'Q', c='blue')
ax.text(K[0], K[1]+0.5, 'K', c='blue')
# ax.scatter([P[0], Q[0]], [P[1], Q[1]])
ax.arrow(P[0], P[1], v[0], v[1], fc='red', ec='red', head_width=head_size, head_length=head_size) # v=PQ
ax.arrow(P[0], P[1], w[0], w[1], fc='red', ec='red', head_width=head_size, head_length=head_size) # w=PK
ax.arrow(P[0], P[1], v_proj[0], v_proj[1], fc='green', ec='green', head_width=head_size, head_length=head_size) # v_proj
ax.arrow(P[0], P[1], w_proj[0], w_proj[1], fc='blue', ec='blue', head_width=head_size, head_length=head_size) # v_proj
ax.grid()
```

### Line
#### Parametric equation
- Equation: $l(t) = p + tv$, where
    - $p$ is a given 2D point that line passes through
    - $v$ is the direction vector of the line
    - $t \in \mathbb{R}$ is the paramter of the line

#### Line segment:
- Equation: $l(t) = (1-t)p + tq$, where
    - $p, q$ are two given 2D points that line passes through
    - $t \in [0,1]$ is the paramter of the line

#### Implicit equation
For any point $q=[x,y]^T$ on the line: $n\cdot(q-p) = 0$
- Where $p=[p_x, p_y]^T$ is the point on the line
- Vector $n=[a, b]^T$ is perpendicular to the line
- $(a,b)^T\cdot(x-p_x, y-p_y)=0$
- $ax + by - (a.p_x + b.p_y) = 0$
- $ax + by + c = 0$ where $c= - (a.p_x + b.p_y)$

```python
import numpy as np
from numpy import array
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

def setup_plot(nrow=1, ncol=1, lim=8):
    plt.rcParams["figure.figsize"] = (5*ncol, 5*nrow)
    fig, ax = plt.subplots(nrow, ncol)
    if ncol > 1:
        for i in range(ncol):
            ax[i].set_xlim([-lim/2, lim]); ax[i].set_ylim([-lim/2, lim]); ax[i].grid()
            ax[i].grid(which='major', color='#CCCCCC', linestyle='--')
            ax[i].grid(which='minor', color='#CCCCCC', linestyle=':')
            ax[i].xaxis.set_major_locator(MultipleLocator(1))
            ax[i].yaxis.set_major_locator(MultipleLocator(1))
    else:
        ax.set_xlim([-lim/2, lim]); ax.set_ylim([-lim/2, lim]); ax.grid()
        ax.grid(which='major', color='#CCCCCC', linestyle='--')
        ax.grid(which='minor', color='#CCCCCC', linestyle=':')
        ax.xaxis.set_major_locator(MultipleLocator(1))
        ax.yaxis.set_major_locator(MultipleLocator(1))
    return ax

```

```python
# Parametric Line
p = np.array([-1,1]).reshape((2,1)) # 2x1
v = np.array([1,1]).reshape((2,1)) # 2x1
n = 100
t = np.linspace(-10, 10, n) # [n=5]
p = np.tile(p, (1, n)) # 2xn
v = np.tile(v, (1, n)) # 2xn
l =  p + t*v

ax = setup_plot(1, 1, lim=12)
ax.text(p[0][0], p[1][0]+0.5, 'P', c='blue')
ax.scatter(p[0], p[1], s=10) # P
ax.scatter(l[0], l[1], s=1) # Lines
```

```python
# Parametric Segment
p = np.array([1, 1]).reshape((2,1)) # 2x1
q = np.array([8, 4]).reshape((2,1)) # 2x1
n= 100 #number of points between P and Q
t = np.linspace(0, 1, n)
p = np.tile(p, (1, n)) # 2xn
q = np.tile(q, (1, n)) # 2xn
l =  (1-t)*p + t*q # 2xn
# print(l)

ax = setup_plot(1, 1, lim=14)
ax.text(p[0][0], p[1][0]+0.5, 'P', c='red')
ax.text(q[0][0], q[1][0]+0.5, 'Q', c='red')
ax.scatter(l[0], l[1], s=3) # Line Segment
ax.legend()
```

```python
# Draw line: via p(1,1), with normal vector norm(a=2, b=1)
# Line equation: ax+by+c=0 => y = (-ax-c)/b
p = np.array([1, 1]).reshape((2,1)) # 2x1
norm = np.array([2, 1]).reshape((2,1)) # 2x1 : normal v ector
a, b = norm
print(a,b)
ax = setup_plot(1, 1, lim=8)

n = 11 # number of points on x segments
x = np.linspace(-2, 2, 11)
c = - np.dot(p.reshape((2,)), norm.reshape((2,)))
print(f'c = {c}')
y = (-a*x - c)/b
ax.text(q[0][0], q[1][0]+0.5, 'Q', c='red')
ax.scatter(q[0], q[1], s=10, fc='red') # Q

ax.scatter(x, y, s=3, fc='blue') # Line
print(x)
print(y)
```

### **2D Transformation**

```python
# Scale
ax = setup_plot(1, 3, lim=10)

def get_scale_matrix(sx, sy):
    return np.array([[sx, 0.0],
                     [0.0, sy]])

xy = np.vstack([x, y]); print(xy.shape)
xy_ret1 = get_scale_matrix(0.5, 0.5)@xy # 
xy_ret2 = get_scale_matrix(0.5, 2.0)@xy # 
ax[0].plot(xy[0], xy[1]) # Original
ax[1].plot(xy_ret1[0], xy_ret1[1]) # Scale1
ax[2].plot(xy_ret2[0], xy_ret2[1]) # Scale2
```

```python
# Shearing
ax = setup_plot(1, 4, lim=8)

def get_shearing_matrix(shx, shy):
    return np.array([  [1.0, shx],
                        [shy, 1.0]])

xy_ret1 = get_shearing_matrix(-0.5, 0)@xy # 
xy_ret2 = get_shearing_matrix(0, -0.5)@xy # 
xy_ret3 = get_shearing_matrix(0.5, 0.5)@xy # 
ax[0].plot(xy[0], xy[1]) # Original
ax[1].plot(xy_ret1[0], xy_ret1[1]) # shearing1
ax[2].plot(xy_ret2[0], xy_ret2[1]) # shearing2
ax[3].plot(xy_ret3[0], xy_ret3[1]) # shearing x,y
```

```python
# Rotation
ax = setup_plot(1, 4, lim=8)

from math import pi, sin, cos
def deg2rad(deg):
    return (deg/180.0)*pi

def get_rotation_matrix(deg):
    rad = deg2rad(deg)
    return np.array([   [cos(rad), -sin(rad)],
                        [sin(rad), cos(rad)]])


xy_ret1 = get_rotation_matrix(10)@xy
xy_ret2 = get_rotation_matrix(20)@xy 
xy_ret3 = get_rotation_matrix(45)@xy

ax[0].plot(xy[0], xy[1]) # Original
ax[1].plot(xy_ret1[0], xy_ret1[1]) # rotation 1
ax[2].plot(xy_ret2[0], xy_ret2[1]) # rotation 2

# Multiple rotation on the same figure
ax[3].plot(xy_ret1[0], xy_ret1[1], label='rot1')
ax[3].plot(xy_ret2[0], xy_ret2[1], label='rot2')
ax[3].plot(xy_ret3[0], xy_ret3[1], label='rot3')
ax[3].legend()
```

```python
# Reflection
ax = setup_plot(1, 1, lim=10)

def get_reflex_matrix(axis):
    if axis == 0: #x-axis
        return np.array([ [1.0, 0.0],
                          [0.0, -1.0]])
    else:
        return np.array([ [-1.0, 0.0],
                          [0.0, 1.0]])

xy_ret1 = get_reflex_matrix(0)@xy # 
xy_ret2 = get_reflex_matrix(1)@xy # 
# Plot on the same chart to compare
ax.plot(xy[0], xy[1], label='Orginal') # Original
ax.plot(xy_ret1[0], xy_ret1[1], label='Reflection x-axis', linestyle='--') 
ax.plot(xy_ret2[0], xy_ret2[1], label='Reflection y-axis', linestyle='--') 
ax.legend()
```

```python
# Chain of transformations

ax = setup_plot(1, 1, lim=12)

xy_ret1 = get_reflex_matrix(1)@xy # Reflect y-axis
xy_ret2 = get_rotation_matrix(30)@xy_ret1 # Rotation
xy_ret3 = get_shearing_matrix(0, 0.5)@xy_ret2 # Shearing

ax.plot(xy[0], xy[1], label='Orginal') # Original
ax.plot(xy_ret1[0], xy_ret1[1], label='Reflection y-axis', linestyle='--') 
ax.plot(xy_ret2[0], xy_ret2[1], label='Rotation 30', linestyle='--') 
ax.plot(xy_ret3[0], xy_ret3[1], label='Shearing', linestyle='--') 
ax.legend()
```

```python
# Chain of transformations 2: Diamon Maker

ax = setup_plot(1, 1, lim=12)

xy_ret1 = get_rotation_matrix(45)@xy # Rotation
xy_ret2 = get_scale_matrix(0.5, 2.0)@xy_ret1 # Scale

ax.plot(xy[0], xy[1], label='Orginal') # Original
ax.plot(xy_ret1[0], xy_ret1[1], label='Rotation 45', linestyle='--') 
ax.plot(xy_ret2[0], xy_ret2[1], label='Scale', linestyle='--') 
ax.legend()
```

### Homogogeous Transforamtion and Homogogeous coordinates
#### Homogogeous Transformation 
- Transformation Matrix $2 \times 2$ can perform scaling, shearing, rotating, reflecting the shape in 2D (so call $R$ transformations), but cannot perform translation in 2D. 
- The homogenous transformation matrix $3 \times 3$ combing above R transformations (size $2 \times 2$) with translation vector $t=[t_x, t_y]$ (size $2 \times 1$)
- $T = \begin{bmatrix} R & t \\ 0 & 1  \end{bmatrix}$
- Translation only if $R$ is identity matrix $\begin{bmatrix} 1 & 0 \\ 0 & 1  \end{bmatrix}$
  - Example: $T = \begin{bmatrix} 1 & 0 & t_x \\ 0 & 1 & t_y \\ 0 & 0 & 1   \end{bmatrix}$

#### Homogogeous coordinates
- Map a 2D point $[x,y]^T$ to a 3D point $(x,y,1)$
- Example translation $[x, y]^T$ to $(t_x, t_y)$: 
  - $T[x, y, 1]^T = \begin{bmatrix} 1 & 0 & t_x \\ 0 & 1 & t_y \\ 0 & 0 & 1 \end{bmatrix} \begin{bmatrix} x \\ y \\ 1\end{bmatrix} = \begin{bmatrix} x + t_x \\  y + t_y \\  1  \end{bmatrix}$

```python
# Example
T1 = np.array([[0.5, 0, 0],
               [0, 0.5, 0],
               [0, 0, 1]]) # 3x3 -> Scale 0.5
T2 = np.array([[0.5, 0, 1],
               [0, 0.5, 1],
               [0, 0, 1]]) # 3x3 -> Scale 0.5 then translate (2,2) and s
xy_homo = np.vstack([xy, np.ones(xy.shape[1])]); print(xy_homo.shape) # 3x5
xy_ret1 = T1@xy_homo
xy_ret2 = T2@xy_homo

ax = setup_plot(1, 1, lim=6)
ax.plot(xy[0], xy[1], label='Orginal') # Original
ax.plot(xy_ret1[0], xy_ret1[1], label='Trasnaltion T1', linestyle='--') 
ax.plot(xy_ret2[0], xy_ret2[1], label='Apply T2 on orginal', linestyle='--')
ax.legend()
```

### Object Center
- Note above transformations are calculated to origin
- If we want to scale/shear/rotate/reflect at objects' centers, we need to move origin to objects' centers

```python
# Orginal
x = np.array([0, 0, 4, 4, 0])
y = np.array([0, 4, 4, 0, 0])

xc = x[:-1].mean(); print(xc) # get all elements in array expect the last one slicing [:-1]
yc = x[:-1].mean(); print(yc)

ax = setup_plot()
ax.plot(x,y)
ax.scatter([xc], [yc], s=10)
```

```python
# Traslate square its center to the origin
T = np.eye(3)
T[:2,2] = np.array([-xc, -yc])

xy_homo = np.vstack([x, y, np.ones(x.shape[0])]); print(xy_homo.shape) # 3x5
new_xy_homo = T@xy_homo
print(xy_homo)
ax = setup_plot(1,1,10)
ax.scatter([xc], [yc], s=10)
ax.plot(x,y, c='blue', label='Original')
ax.plot(new_xy_homo[0], new_xy_homo[1], c='red', label='Translated')


T = np.eye(3)
T[:2,:2] = get_rotation_matrix(45)
xy_rotation = T@new_xy_homo
ax.plot(xy_rotation[0], xy_rotation[1], c='red', label='Translated-Rotated', linestyle='--')

# Move the square back to its center
T = np.eye(3)
T[:2,2] = np.array([xc, yc])
xy_rotation_center = T@xy_rotation
ax.plot(xy_rotation_center[0], xy_rotation_center[1], c='green', label='Translated-Rotated-Center', linestyle='--')

ax.legend()
```
